### Круговая диаграмма расходов в течении недели.

Азот – 78%

Кислород – 21%

Другие газы – 1%

```python
import matplotlib
import numpy as np
vals = [78, 21, 1]

labels = ["Озот", "Кислород", "Другие газы"]
explode = (0.1, 0.1, 0.1)

fig, ax = plt.subplots()

ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode,
wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, rotatelabels=True)
ax.axis("equal")
```

#### С противоположных концов страны длиной 120 км едут навстречу друг другу
#### два автомобиля. Через сколько секунд они встретятся, если второй начнет ехать через 5 часов
#### после первого и если первый едет со скоростью 6 км/ч, а второй – 9 км/ч? [2].

#### РЕШЕНИЕ
Пусть х ч.– время движения второго автомобиля, тогда время первого – (х+5) ч. Составим
уравнение, учитывая, что сумма расстояний равна 120 км:6(х + 5) + 9х = 120 решив которое, получим, что х=
6 ч. Тогда время, через которое они встретятся – 11 ч.

```python
import matplotlib.pyplot as plt
plt.plot([120, 100,80, 60, 40, 20,10])
plt.plot([0, 5, 10, 15, 18, 20, 25])
plt.ylabel('x, м', fontsize=12, color='blue')
plt.xlabel('t, с', fontsize=12, color='blue')
plt.text(4.8, 270, '')
plt.text(4.8, 35, '')
plt.grid()
plt.show()
```
